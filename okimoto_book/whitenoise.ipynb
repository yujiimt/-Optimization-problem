```
import numpy as np
import random
import matplotlib.pyplot as plt

%matplotlib inline
```

```
#ノイズデータのデータ長
dlen = 1024
#ノイズの平均
mean = 0.0
#ノイズの分散
std = 1.0


y = np.array([random.gauss(mean, std) for i in range(dlen)])

plt.figure(figsize=(12, 6))
plt.plot(y)
plt.show()
```
